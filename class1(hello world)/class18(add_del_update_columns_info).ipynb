```python
# asi data pe grade lgne k liye function bnane bne ga

def grade(per:int)->str:
    if per >= 80:
        return "A+"
    elif per >= 70:
        return "A"
    elif per >= 60:
        return "B"
    elif per >= 50:
        return "C"
    elif per >= 40:
        return "D"
    elif per >= 33:
        return "E"
    else:
        return "F"

grade(99)
```

```python
# function use kr k obtain col bnya
def fn_obtained(s1: int, s2: int,s3: int,s4: int,s5: int)->int:
    return s1+s2+s3+s4+s5

df['obtained'] = df[["s1", "s2", "s3", "s4", "s5"]].apply(lambda x:fn_obtained(*x), axis=1) 
df
```

```python
# Multi output multi column
def fn_obtained(s1: int, s2: int,s3: int,s4: int,s5: int)->tuple[int,int,float,str]:
    total: int = 500
    obtained : int = s1+s2+s3+s4+s5
    per : float = obtained/total * 100
    grade : str = ""
    if per >= 80:
        grade = "A+"
    elif per >= 70:
        grade = "A"
    elif per >= 60:
        grade = "B"
    elif per >= 50:
        grade = "C"
    elif per >= 40:
        grade = "D"
    elif per >= 33:
        grade = "E"
    else:
        grade = "F"
    return total, obtained, per, grade

df[["Total",'obtained','percentage',"Grade"]] = df[["s1", "s2", "s3", "s4", "s5"]].apply(lambda x:fn_obtained(*x), axis=1,result_type="expand") 
df
```

# Map function for remaks

```python
from typing import Dict

remark : Dict[str,str] = {
    "A+" : "Excellent",
    "A" : "Very Good",
    "B" : "Good",
    "C" : "Fair",
    "D" : "Poor",
    "E" : "Very Poor",
    "F" : "Fail",
}
df["Remark"] = df.Grade.map(remark)
df
```
